## <font color='#4287f5'>Before we start, lets know about the disease which we are working inüòá... </font>

**Diabetes mellitus(DM)** is a disease condition which comprises a group of metabolic disorders that share 
the common feature of **hyperglycemia** (higher blood glucose concentration).

### <font color='#4287f5'>How Diabetes is classified? </font>
DM is currently classified on the basis  of the pathogenic process that leads to hyperglycemia. According to which,we have two types of DM:

1.<u>**Type 1 DM**</u> : It is characterized by **insulin deficiency**ü•µ commonly caused by autoimmune destruction of the pancreatic islet beta cells (Autoimmune disease,genetic cause).[***Auto: self ,immune: immunity i.e, self destruction of our own defence system***]

## <font color='#4287f5'>Lets now look at the risk factorsüòü for DM... </font>

### <font color='#22225B'>(Knowing this, might help us in selecting features for our model üòå...) </font>

* <u>**For type I**</u>:
    -genetic variations,inborn error of metabolism (**Factors which are not under our control**)
    
* <u>**For type II**</u>:
    1. Family history of diabetes (i.e., parent or sibling with type 2 diabetes)
    2. Overweight or obese (BMI ‚â•25 kg/m2, ‚â•23 kg/m2 in Asian Americans, or other ethnically relevant definition for overweight)
    3. Physical inactivity
    4. Race/ethnicity (e.g., African American, Latino, Native American, Asian American, Pacific Islander)
    5. Previously identified with IFG, IGT, or a hemoglobin A1c of 5.7‚Äì6.4%
    6. History of GDM
    7. Hypertension (blood pressure ‚â•140/90 mmHg)
    8. HDL cholesterol level <35 mg/dL (0.90 mmol/L) and/or a triglyceride level >250 mg/dL (2.82 mmol/L)
    9. Polycystic ovarian syndrome(in women)
    10. History of cardiovascular disease
    
As clear from the risk factors for type II disease, it is definitely possible to regulate some of the factors which lead to disease,.....(**This is the area of focus because if we know these factors, it will help us during feature engineering phase of creating AI model**)
