### Lets talk about the way blood reports are generated?

* Blood reports take time to get generated. The reason behind is that different tests take different time to come up with the results. Most in-hospital results can be obtained within three to six hours after taking the blood. Sometimes blood drawn at other,non-hospital facilities can take several days to get results.
* Location can play a factor into how fast you will get your results back. For example, going to a place where there's an on-site laboratory(such as a hospital) may get you results more quickly than if your blood has to be sent to another laboratory. Specialty tests for rare conditions often need to be sent to specific laboratories.
* Somtimes,how quickly the blood tests take depends on how common the blood test is. Blood tests performed more often,such as a CBC or metabolic panel, are usually available more quickly than tests for rare conditions.
* With innovations in quick testing, many more laboratory tests are available sooner than ever before. However, it's often important that your doctor do a careful review before passing along the results. Asking a doctor or laboratory technicians about how long average tests will take can help you to establish a realistic time frame for getting results.

Lets first insert the 'ID','age' and 'sex' columns to the dataset.

```python
ID = list(np.linspace(0,1000,1000).astype(int))
```

```python
age = np.random.normal(50, 15, 1000).astype(int).tolist()
age = [(x + 10) if x < 5 else x for x in age]   # I dont want to include the cases who are less than 5 years old...
```

```python
# 1 - male
# 0 - female
sex = []
for i in range(0,1000):
    x = random.randint(0, 1)
    sex.append(x)
```

```python
Anemic_status = []
for i in range(0,1000):
    x = random.randint(0, 3)
    Anemic_status.append(x)
    
```

```python
# create a DataFrame
Report = {'ID': ID,'age': age,'sex': sex,'Anemic_status': Anemic_status}
data = pd.DataFrame(Report)
```

```python
data
```

```python
data.insert(4,'Hb_level',True)
```

```python
def get_Hb_level(x):
    if x == 0:
        return round(random.uniform(13,16),2)
    elif x == 1:
        return round(random.uniform(9,13),2)
    elif x == 2:
        return round(random.uniform(6,9),2)
    else:
        return round(random.uniform(2,6),2)
```

```python
data['Hb_level'] = data['Anemic_status'].apply(lambda x: get_Hb_level(x))
```

```python
data.insert(5,'RBC_count',True)
```

```python
# RBC_count in (value x 10^12/L)
def f(x):
    if x['sex'] == 0 and x['Anemic_status'] == 0:
        return round(random.uniform(3.5,5.0),2)
    elif x['sex'] == 0 and x['Anemic_status'] == 1:
        return round(random.uniform(3.2,4.8),2)
    elif x['sex'] == 0 and x['Anemic_status'] == 2:
        return round(random.uniform(2.5,3.2),2)
    elif x['sex'] == 0 and x['Anemic_status'] == 3:
        return round(random.uniform(1.8,2.5),2)
    elif x['sex'] == 1 and x['Anemic_status'] == 0:
        return round(random.uniform(4.5,5.5),2)
    elif x['sex'] == 1 and x['Anemic_status'] == 1:
        return round(random.uniform(4.0,4.8),2)
    elif x['sex'] == 1 and x['Anemic_status'] == 2:
        return round(random.uniform(3.0,4.0),2)
    else:
        return round(random.uniform(2.0,3.0),2)
```

```python
data['RBC_count'] = data.apply(f, axis=1)
```

```python
data.insert(6,'MCV',True)
```

```python
# MCV in fL
def get_MCV(x):
    if x == 0:
        return random.randint(80,100)
    elif x == 1:
        return random.randint(75,95)
    elif x == 2:
        return random.randint(70,90)
    else:
        return random.randint(65,85)
```

```python
data['MCV'] = data['Anemic_status'].apply(lambda x: get_MCV(x))
```

```python
data.insert(7,'HCT',True)
```

```python
data['HCT'] = (data['RBC_count'].multiply(data['MCV']))/10
```

```python
data['HCT'] = round(data['HCT'],2)
```

```python
data.insert(8,'MCH',True)
```

```python
data['MCH'] = (data['Hb_level'].multiply(10))/data['RBC_count']
```

```python
data['MCH'] = round(data['MCH'],2)               # normal range : 26-36pg
```

```python
data.insert(9,'MCHC',True)
```

```python
data['MCHC'] = (data['Hb_level'].multiply(100))/data['HCT']
```

```python
data['MCHC'] = round(data['MCHC'],2)               # normal range : 30-36%
```

```python
data.insert(10,'WBC',True)
```

```python
def get_WBC(x):
    return round(random.uniform(3,19),1)
```

```python
data['WBC'] = data['WBC'].apply(lambda x: get_WBC(x))
```

```python
data['Neutrophils'] = (data['WBC'].multiply(round(random.uniform(50,70),2))/100)
```

```python
data['Neutrophils'] = round(data['Neutrophils'],2)
```

```python
data['Lymphocytes'] = (data['WBC'].multiply(round(random.uniform(20,40),2))/100)
```

```python
data['Lymphocytes'] = round(data['Lymphocytes'],2)
```

```python
data['Monocytes'] = (data['WBC'].multiply(round(random.uniform(5,10),2))/100)
```

```python
data['Monocytes'] = round(data['Monocytes'],2)
```

```python
data['Eosinophils'] = (data['WBC'].multiply(round(random.uniform(0,2),2))/100)
```

```python
data['Eosinophils'] = round(data['Eosinophils'],2)
```

```python
data['Basophils'] = (data['WBC'].multiply(round(random.uniform(2,4),2))/100)
```

```python
data['Basophils'] = round(data['Basophils'] ,2)
```

```python
data.insert(16,'Platelet_count',True)
```

```python
Platelet_count = []
for i in range(0,1000):    
    x = round(random.uniform(0.4,6),2)     # normal range of platelets 1.5 - 4 Lakhs/L
    Platelet_count.append(x)
    
```

```python
data['Platelet_count'] = Platelet_count
```

```python
Sodium_level = []
for i in range(0,1000):    
    x = random.randint(125,155)     # normal range 125 - 145 mmol/L
    Sodium_level.append(x)
    
Potassium_level = []
for i in range(0,1000):    
    x = round(random.uniform(2.8,5.5),1)    # normal range 3.6 - 5.2 mmol/L
    Potassium_level.append(x)
    
Chloride_level = []
for i in range(0,1000):    
    x = random.randint(86,115)     # normal range 96 - 106 mmol/L
    Chloride_level.append(x)
    
Bicarbonate_level = []
for i in range(0,1000):    
    x = random.randint(16,35)     # normal range 22 - 29 mmol/L
    Bicarbonate_level.append(x)

Blood_urea_level = []
for i in range(0,1000):    
    x = round(random.uniform(1.2,9.0),1)    # normal range 2.1 - 8.5 mmol/L
    Blood_urea_level.append(x)

Serum_creatinine_level = []
for i in range(0,1000):    
    x = random.randint(55,120)     # normal range 66 - 112 micro-mol/L
    Serum_creatinine_level.append(x)  
```

```python
data['Sodium_level'] = Sodium_level
data['Potassium_level'] = Potassium_level
data['Chloride_level'] = Chloride_level
data['Bicarbonate_level'] = Bicarbonate_level
data['Blood_urea_level'] = Blood_urea_level
data['Serum_creatinine_level'] = Serum_creatinine_level
```
