```python
import pandas as pd                       # library for data manipulation (data cleaning,analysing...)
import numpy as np                        # library for numerical computation
import matplotlib.pyplot as plt           # library for data visualization which is built on NumPy arrays
%matplotlib inline
import seaborn as sns                     # library for data visualization built on top of matplotlib and 
                                          # closely integrated with pandas data structures in Python.

import random

import warnings                           # to avoid warning flash
warnings.filterwarnings('ignore')

# Other libraries will be imported in the following steps as the process moves.... 
```

```python
ID = list(np.linspace(0,1000,1000).astype(int))
```

```python
age = np.random.normal(50, 15, 1000).astype(int).tolist()
age = [(x + 20) if x < 10 else x for x in age]
```

```python
sex = []
for i in range(0,1000):
    x = random.randint(0, 1)               # 0 - female    1 - male
    sex.append(x)
```

```python
Anemic_status = []
for i in range(0,1000):
    x = random.randint(0, 3)
    Anemic_status.append(x)
    
```

```python
# create a DataFrame
Report = {'ID': ID,'age': age,'sex': sex,'Anemic_status': Anemic_status}
df = pd.DataFrame(Report)
```

```python
df.insert(4,'Hb_level',True)
```

```python
def get_Hb_level(x):
    if x == 0:
        return round(random.uniform(13,16),2)
    elif x == 1:
        return round(random.uniform(9,13),2)
    elif x == 2:
        return round(random.uniform(6,9),2)
    else:
        return round(random.uniform(2,6),2)
```

```python
df['Hb_level'] = df['Anemic_status'].apply(lambda x: get_Hb_level(x))
```

```python
df
```

```python
df.insert(5,'managing_anemia',True)
```

```python
def get_managing_anemia(x):
    if x == 0:
        return 0                         # 0 - continue healthy lifestyle with balanced diet, moderate exercise and good sleep
    elif x == 1:
        return 1                         # 1 - mild anemia can be managed conservatively by prescribing iron rich foods or oral tabs
    elif x == 2:
        return 2                         # 2 - oral tabs + Inj, evaluate for the cause and treat it...
    else:
        return 3                         # 3 - Hospitalization and blood transfusion,evaluate for the cause and treat it..
```

```python
df['managing_anemia'] = df['Anemic_status'].apply(lambda x: get_managing_anemia(x))
```

```python
df
```

```python
df.insert(6,'RBC_count',True)
```

```python
# RBC_count in (value x 10^12/L)
def f(x):
    if x['sex'] == 0 and x['Anemic_status'] == 0:
        return round(random.uniform(3.5,5.0),2)
    elif x['sex'] == 0 and x['Anemic_status'] == 1:
        return round(random.uniform(3.2,4.8),2)
    elif x['sex'] == 0 and x['Anemic_status'] == 2:
        return round(random.uniform(2.5,3.2),2)
    elif x['sex'] == 0 and x['Anemic_status'] == 3:
        return round(random.uniform(1.8,2.5),2)
    elif x['sex'] == 1 and x['Anemic_status'] == 0:
        return round(random.uniform(4.5,5.5),2)
    elif x['sex'] == 1 and x['Anemic_status'] == 1:
        return round(random.uniform(4.0,4.8),2)
    elif x['sex'] == 1 and x['Anemic_status'] == 2:
        return round(random.uniform(3.0,4.0),2)
    else:
        return round(random.uniform(2.0,3.0),2)
    
    
```

```python
df['RBC_count'] = df.apply(f, axis=1)
```

```python
def get_infection_status(x):
    if x < 4:
        return 0                        # leucocytopenia
    elif x >= 4 and x < 11:
        return 1                        # normal
    else:
        return 2                        # Leucocytosis
```

```python
df.insert(18,'managing_infection',True)
```

```python
def how_to_manage_infection(x):
    if x == 0:
        return 0        # 0 - evaluate for the low wbc count i.e low immunity, check for HIV,DM or other immunocompromised state
    elif x == 1:
        return 1        # 1 - normal, nothing to worry
    else:
        return 2        # 2 - '!!!there is infection',find the foci of infection and treat the cause with suitable antibiotics
```

```python
df['managing_infection'] = df['infection_status'].apply(lambda x: how_to_manage_infection(x))
```

```python
df.insert(19,'Platelet_count',True)
```

```python
Platelet_count = []
for i in range(0,1000):    
    x = round(random.uniform(0.4,6),2)     # normal range of platelets 1.5 - 4 Lakhs/L
    Platelet_count.append(x)
    
```

```python
df['Platelet_count'] = Platelet_count
```

```python
df[df['Platelet_count'] < 1.5].value_counts().sum()
```

```python
conditions = [
    df['Platelet_count'] < 1.5,
    (df['Platelet_count'] >= 1.5) & (df['Platelet_count'] < 5 ),
    df['Platelet_count'] >= 5
]

choices = [0,1,2]         # 0- Thrombocytopenia     1-normal     2-thrombocytosis

df['Bleeding_status'] = np.select(conditions, choices, default=0)
```

```python
df
```

```python
df['Bleeding_status'].value_counts()
```

```python
#now manage platelet disorders......
def how_to_manage_platelet_disorder(x):
    if x == 0:       # Thrombocytopenia
        return 0     # evaluate for the cause and if critical, transfuse platelet precipitates or FFP's
    elif x == 1:
        return 1     # normal......nothing to worry
    else:
        return 2     # Thrombocytosis, evaluate for the cause and treat it ....
```

```python
df['managing_platelet_disorders'] = df['Bleeding_status'].apply(lambda x: how_to_manage_platelet_disorder(x))
```

```python
df.insert(32,"Conjugated_Bilirubin_count",True)
```

```python
def get_Conjugated_Bilirubin_count(x):
    if x == 0:
        return round(random.uniform(0.1,0.3),2)   # normal
    elif x == 1:
        return round(random.uniform(0.1,0.4),2)
    elif x == 2:
        return round(random.uniform(0.4,3.0),2)
    else :
        return round(random.uniform(3,15),2)
```

```python
df['Conjugated_Bilirubin_count'] = df['LFT_Diagnosis'].apply(lambda x : round(get_Conjugated_Bilirubin_count(x),2))
                
```

```python
df.insert(33,"Unconjugated_Bilirubin_count",True)
```

```python
df.insert(34,"Total_Bilirubin_count",True)
```

```python
def get_Total_Bilirubin_count(x):
    if x == 0:
        return round(random.uniform(0.1,1.0),2)
    if x == 1:
        return round(random.uniform(2.5,12),2)
    elif x == 2:
        return round(random.uniform(5,12),2)
    else :
        return round(random.uniform(16,18),2)
```

```python
df['Total_Bilirubin_count'] = df['LFT_Diagnosis'].apply(lambda x : round(get_Total_Bilirubin_count(x),2))
                
```

```python
df['Unconjugated_Bilirubin_count'] = df['Total_Bilirubin_count'] - df['Conjugated_Bilirubin_count']
```
