### What is the target variable in this dataset?

The target variable is not yet inserted yet to the dataset !!!
In this dataset, I would like to insert as many target variables as possible so that I can prepare many AI models out of it....

```python
# first target variable -- 'Anemic_status'
```

```python
df.insert(4,'Anemic_status',True)
```

```python
def get_Anemic_status(x):
    if x >= 12.5:
        return 0       # normal
    elif x >= 9 and x < 12.5:
        return 1       # mild anaemia
    elif x >= 6 and x < 9:
        return 2       # moderate anaemia
    else:
        return 3       # severe anaemia
```

```python
df['Anemic_status'] = df['Hb_level'].apply(lambda x: get_Anemic_status(x))
```

```python
df
```

```python
df.insert(5,'managing_anemia',True)
```

```python
def get_managing_anemia(x):
    if x == 0:
        return 0                         # 0 - continue healthy lifestyle with balanced diet, moderate exercise and good sleep
    elif x == 1:
        return 1                         # 1 - mild anemia can be managed conservatively by prescribing iron rich foods or oral tabs
    elif x == 2:
        return 2                         # 2 - oral tabs + Inj, evaluate for the cause and treat it...
    else:
        return 3                         # 3 - Hospitalization and blood transfusion,evaluate for the cause and treat it..
```

```python
df['managing_anemia'] = df['Anemic_status'].apply(lambda x: get_managing_anemia(x))
```

```python
df
```

```python
# 2nd target variable -- 'infection_status'
```

```python
df.insert(17,'infection_status',True)
```

```python
def get_infection_status(x):
    if x < 4:
        return 0                        # leucocytopenia
    elif x >= 4 and x < 11:
        return 1                        # normal
    else:
        return 2                        # Leucocytosis
```

```python
df['infection_status'] = df['WBC'].apply(lambda x: get_infection_status(x))
```

```python
df.insert(18,'managing_infection',True)
```

```python
def how_to_manage_infection(x):
    if x == 0:
        return 0        # 0 - evaluate for the low wbc count i.e low immunity, check for HIV,DM or other immunocompromised state
    elif x == 1:
        return 1        # 1 - normal, nothing to worry
    else:
        return 2        # 2 - '!!!there is infection',find the foci of infection and treat the cause with suitable antibiotics
```

```python
df['managing_infection'] = df['infection_status'].apply(lambda x: how_to_manage_infection(x))
```

```python
df
```

```python
# 3rd target variable -- 'Bleeding_status'
```

```python
df.insert(20,'Bleeding_status',True)
```

```python
conditions = [
    df['Platelet_count'] < 1.5,
    (df['Platelet_count'] >= 1.5) & (df['Platelet_count'] < 5 ),
    df['Platelet_count'] >= 5
]

choices = [0,1,2]         # 0- Thrombocytopenia     1-normal     2-thrombocytosis

df['Bleeding_status'] = np.select(conditions, choices, default=0)
```

```python
df.insert(21,'managing_platelet_disorders',True)
```

```python
#now manage platelet disorders......
def how_to_manage_platelet_disorder(x):
    if x == 0:       # Thrombocytopenia
        return 0     # evaluate for the cause and if critical, transfuse platelet precipitates or FFP's
    elif x == 1:
        return 1     # normal......nothing to worry
    else:
        return 2     # Thrombocytosis, evaluate for the cause and treat it ....
```

```python
df['managing_platelet_disorders'] = df['Bleeding_status'].apply(lambda x: how_to_manage_platelet_disorder(x))
```

```python
df
```

```python
# 4th target variable

df.insert(23,'Sodium_level_status',True)

def get_Sodium_level_status(x):
    if x < 125:
        return 0        # hyponatraemia(correction needed)
    elif x >145:
        return 2        # hypernatraemia (correction needed)
    else:
        return 1        # normal
    
df['Sodium_level_status'] = df['Sodium_level'].apply(lambda x:get_Sodium_level_status(x))
```

```python
# 5th target variable

df.insert(25,'Potassium_level_status',True)

def get_Potassium_level_status(x):
    if x < 3.5:
        return 0        # hypokalaemia(correction needed)
    elif x >5:
        return 2        # hyperkalaemia (correction needed)
    else:
        return 1        # normal
    
df['Potassium_level_status'] = df['Potassium_level'].apply(lambda x: get_Potassium_level_status(x))
```

```python
# 6th target variable

df.insert(27,'Chloride_level_status',True)

def get_Chloride_level_status(x):
    if x < 96:
        return 0        # hypochloraemia(correction needed)
    elif x >106:
        return 2        # hyperchloraemia (correction needed)
    else:
        return 1        # normal
    
df['Chloride_level_status'] = df['Chloride_level'].apply(lambda x:get_Chloride_level_status(x))
```

```python
# 7th target variable

df.insert(29,'Bicarbonate_level_status',True)

def get_Bicarbonate_level_status(x):
    if x < 96:
        return 0        # hypochloraemia(correction needed)
    elif x >106:
        return 2        # hyperchloraemia (correction needed)
    else:
        return 1        # normal
    
df['Bicarbonate_level_status'] = df['Bicarbonate_level'].apply(lambda x:get_Bicarbonate_level_status(x))
```

```python
df
```

```python
# 8th target variable

df.insert(32,'kidney_status',True)

def get_kidney_status(x):
    if x > 2:
        return 1        # unhealthy (high S.Creat level indicates kidney pathology which may be either acute or chronic)
    else:
        return 0        # healthy status

df['kidney_status'] = df['Serum_creatinine_level'].apply(lambda x: get_kidney_status(x))    
```

```python
# 9th target variable

df.insert(36,"Diabetic_status",True)

def get_Diabetic_status(x):
    if x > 6.5:                  # if the HbA1c values > 6.5 , then Diabetic
        return 2
    elif x > 5.7 and x < 6.5:    # if the HbA1c values between 5.7 and 6.5 , then Pre-Diabetic
        return 1
    else :                       # # if the HbA1c values < 5.7 , then normal.
        return 0
    
# Set a length of the list to 1000
for i in range(0, 1000):
    df['Diabetic_status'] = df['A1C'].apply(lambda x: get_Diabetic_status(x))
                
```

```python
# 10th target variable

df.insert(45,"LFT_status",True)

# 0 ---> Normal
# 1 ---> Jaundice

def get_LFT_status(x):
    if x > 0 and x < 1:                 
        return 0
    else :                       
        return 1

df['LFT_status'] = df['Total_Bilirubin_count'].apply(lambda x: get_LFT_status(x))    
```
