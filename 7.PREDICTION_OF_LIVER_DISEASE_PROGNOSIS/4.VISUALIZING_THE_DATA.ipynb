## <font color='#4287f5'> Lets make a new target variable.....</font>

Before that, lets know something called **"CHILD-PUGH CLASSIFICATION"**

**What is Child-Pugh classification in cirrhosis ?**
*  The Child-Pugh (CP) scoring classification, originally devised to risk-stratify patients undergoing shunt surgery is useful to assess liver disease severity in patients with established cirrhosis
* In cirrhosis, Child‚Äôs grade is used **to assess the hepato-cellular function**. 
* The determining 5 factors are :\
    * serum bilirubin
    * serum albumin
    * ascites
    * hepatic encephalopathy
    * prothrombin time. 
* Every patient is assigned a class (A, B and C). **Class A is ‚Äôprognostically‚Äô better** and **class C is the worst**. Child-Pugh scoring system (scores 5-15 encompassing class A, B, C) is used to assess prognosis and is a reliable predictor of survival.

Out of the above 5 features, we have 4 features in our dataset. We don't have the feature on encephalopathy. So, instead of that feature, we can include the 'Histologic_stage_feature' to classify the severity.Lets do it...

```python
# Lets define functions to get points for the above 5 features mentioned....

def get_Ascites_point(x):
    if x == 'Y':
        return 3
    elif x == 'S':
        return 2
    else:
        return 1
                

def get_Albumin_point(x):
    if x <= 2.8:
        return 3
    elif (x > 2.8) & (x <= 3.5):
        return 2
    else:
        return 1

            
def get_Prothrombin_point(x):
    if x >= 6.0:
        return 3
    elif (x >= 4.0) & (x < 6.0):
        return 2
    else:
        return 1 


def get_Bilirubin_point(x):
    if x >= 3.0:
        return 3
    elif (x >= 2.0) & (x < 3.0):
        return 2
    else:
        return 1
            
def get_Stage_point(x):
    if x == 4.0:
        return 3
    elif (x == 2.0) | (x == 3.0):
        return 2
    else:
        return 1  
        
```

```python
# Lets get total points to assess severity score...
df1['Total_points'] = df1['Edema'].apply(lambda x: get_Ascites_point(x)) + df1['Albumin'].apply(lambda x: get_Albumin_point(x)) + df1['Prothrombin'].apply(lambda x: get_Prothrombin_point(x)) + df1['Bilirubin'].apply(lambda x:get_Bilirubin_point(x)) + df1['Stage'].apply(lambda x:get_Stage_point(x))
```

```python
# Lets get the severity score which gives us prognosis similar to that of Child-Pugh score

def get_severity_score(x):
    if (x > 5) and (x <= 7):
        return 'CLASS A'
    elif (x > 8) and (x <= 10):
        return 'CLASS B'
    else:
        return 'CLASS C'

df1['Severity_class'] = df1['Total_points'].apply(lambda x: get_severity_score(x))
```

```python
# Lets get the percentage of 'Severity_class' class...

def get_percentage(ax, feature):
    total = len(feature)
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() / 2 - 0.05
        y = p.get_y() + p.get_height()
        ax.annotate(percentage, (x, y), size = 12)
```
