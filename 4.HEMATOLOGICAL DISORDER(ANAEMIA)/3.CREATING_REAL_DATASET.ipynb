As I have told before, I want my dataset to look real. So,lets change the values of each features...

```python
# a function to convert reticulocyte count values in the correct format
def convert_retic_count(x):
    if x == 0:                         
        return round(random.uniform(0,0.5),2)
    elif x == 1:           
        return round(random.uniform(0.5,2.5),2)
    else:
        return round(random.uniform(2.5,6.0),2)                        
```

```python
df['Retic_count'] = df['Retic_count'].apply(lambda x: convert_retic_count(x))
```

```python
df
```

```python
# a function to convert bilirubin count values in the correct format
def convert_bilirubin_count(x):
    if x == 3:              #normal
        return round(random.uniform(0.1,0.8),2)
    elif x == 4:
        return round(random.uniform(0.8,8.0),2)                         #high
    else:
        return x
```

```python
df['Bilirubin_count'] = df['Bilirubin_count'].apply(lambda x: convert_bilirubin_count(x))
```

```python
df
```

```python
# a function to convert LDH count values in the correct format
def convert_LDH_count(x):
    if x == 3:              #normal
        return random.randint(5,240)
    elif x == 4:
        return random.randint(240,600)                       #high
    else:
        return x
```

```python
df['LDH_count'] = df['LDH_count'].apply(lambda x: convert_LDH_count(x))
```

```python
df
```

```python
# a function to convert Coombs test values in the correct format
def convert_Coombs_test(x):
    if x == 5:              
        return 'positive'
    elif x == 6:
        return 'negative'                    
    else:
        return x
```

```python
df['Coombs_test'] = df['Coombs_test'].apply(lambda x:convert_Coombs_test(x))
```

```python
df
```

```python
# a function to convert Peripheral smear values in the correct format
def convert_Peripheral_Smear(x):
    if x == 7:              
        return 'microcytic'
    elif x == 8:
        return 'normocytic'
    elif x == 9:
        return 'macrocytic'
    else:
        return x
```

```python
df['Peripheral_Smear'] = df['Peripheral_Smear'].apply(lambda x: convert_Peripheral_Smear(x))
```

```python
df
```

```python
# a function to convert Ferritin_level values in the correct format
def convert_Ferritin_level(x):
    if x == 10:                         #low
        return random.randint(2,18)
    elif x == 11:           #normal(18 to 200ng/ml)
        return random.randint(18,200)
    else:
        return x                    
```

```python
df['Ferritin_level'] = df['Ferritin_level'].apply(lambda x:convert_Ferritin_level(x))
```

```python
df
```

```python
# a function to convert Hb_electrophoresis values in the correct format
def convert_Hb_electrophoresis(x):
    if x == 12:                       
        return 'HbA2,HbF'
    elif x == 13:          
        return 'normal'
    else:
        return x                      
```

```python
df['Hb_electrophoresis'] = df['Hb_electrophoresis'].apply(lambda x: convert_Hb_electrophoresis(x))
```

```python
df
```

```python
# a function to convert blood cell count values in the correct format
def convert_blood_cell_count(x):
    if x == 14:                         
        return 'pancytopenia_present'
    elif x == 15:         
        return 'normal_count'
    else:
        return x                     
```

```python
df['Other_blood_cell_count\(Pancytopenia +/-)'] = df['Other_blood_cell_count\(Pancytopenia +/-)'].apply(lambda x:convert_blood_cell_count(x))
```

```python
df
```

```python
# a function to convert Serum_Creatinine values in the correct format
def convert_Serum_Creatinine(x):
    if x == 16:                         #normal(0.6 to 1.2mg/dl)
        return round(random.uniform(0.6,1.2),1)
    elif x == 17:           
        return round(random.uniform(1.2,10.0),1)
    else:
        return x                      
```

```python
df['Serum_Creatinine'] = df['Serum_Creatinine'].apply(lambda x: convert_Serum_Creatinine(x))
```

```python
df
```

```python
# a function to convert TSH_Level values in the correct format
def convert_TSH_Level(x):
    if x == 18:                         #normal(0.5 to 5.0(¬µU/mL))
        return round(random.uniform(0.5,5.0),2)
    elif x == 19:           
        return round(random.uniform(5.0,10.0),2)
    else:
        return x                      
```

```python
df['TSH_Level'] = df['TSH_Level'].apply(lambda x: convert_TSH_Level(x))
```

```python
df
```

```python
# a function to convert Serum_electrophoresis values in the correct format
def convert_Serum_electrophoresis(x):
    if x == 21:                     
        return 'M-spike'
    elif x == 20:           
        return 'normal'
    else:
        return x                       
```

```python
df['Serum_electrophoresis'] = df['Serum_electrophoresis'].apply(lambda x: convert_Serum_electrophoresis(x))
```

```python
df
```

Next, I convert 'B12 or Folate level' column as 'Vit_B12_level' because folate level testing is rarely done nowdays.

```python
df.rename(columns = {'B12_or_Folate_levels':'Vit_B12_level'},inplace = True)
```

```python
# a function to convert Vit_B12_level values in the correct format
def convert_Vit_B12_level(x):
    if x == 22:                         #low
        return random.randint(10,190)
    elif x == 23:           #normal(190 to 950pg/ml)
        return random.randint(190,950)
    else:
        return x                    
```

```python
df['Vit_B12_level'] = df['Vit_B12_level'].apply(lambda x: convert_Vit_B12_level(x))
```

Now,our dataframe looks like the real one but one thing is so much depressing i.e, most of the feature values are filled with 'not_done' values.
```python
df.loc[rows,'Other_blood_cell_count\(Pancytopenia +/-)'] = df['Other_blood_cell_count\(Pancytopenia +/-)'].apply(lambda x: 'normal_count' )
```
