```python
# import the modules
import pandas as pd
import numpy as np
import random
```

```python
ID = list(np.linspace(0,1000,1000).astype(int))
```

```python
age = np.random.normal(50, 15, 1000).astype(int).tolist()
age = [(x + 20) if x < 10 else x for x in age]
```

```python
sex = []
for i in range(0,1000):
    x = random.randint(0, 1)
    sex.append(x)
```

```python
Retic_count = []
for i in range(0,1000):
    x = random.randint(0, 2)
    Retic_count.append(x)
    
```

```python
# create a DataFrame
Report = {'ID': ID,'age': age,'sex': sex,'Retic_count': Retic_count}
df = pd.DataFrame(Report)
```

```python
df
```

```python
df['sex'].value_counts()
```

```python
df.insert(4,"Bilirubin_count",True)
```

```python
def get_Bilirubin_count(x):
    if x == 0 or x == 1:
        return 0
    else :
        return random.randint(0 ,1)
```

```python
# Set a length of the list to 1000
for i in range(0, 1000):
    df['Bilirubin_count'] = df['Retic_count'].apply(lambda x : get_Bilirubin_count(x))
                
```

```python
df
```

```python
df['Bilirubin_count'].value_counts()
```

```python
df.insert(5,"LDH_count",True)
```

```python
def get_LDH_count(x):
    if x == 0 :
        return 0
    else :
        return 1
```

```python
df['LDH_count'] = df['Bilirubin_count'].apply(lambda x: get_LDH_count(x))
```

```python
df
```

```python
df['LDH_count'].value_counts()
```

```python
df.insert(6,"Coombs'_test",True)
```

```python
def get_Coombs_test(x):
    if x == 1:
        return random.randint(0,1)
    else:
        return 0
```

```python
df["Coombs'_test"] = df['LDH_count'].apply(lambda x: get_Coombs_test(x))
```
