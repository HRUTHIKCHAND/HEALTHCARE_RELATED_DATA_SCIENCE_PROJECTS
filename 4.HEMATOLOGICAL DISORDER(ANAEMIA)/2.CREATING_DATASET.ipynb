<h1 style= "text-align:center;color:#052261;font-size:120%;"><u> RETICULOCYTE COUNT  </u></h2>
<h2 style= "color:#38bdc2; font-family:Verdana,Arial; font-size:15pt"> I)What is a reticulocyte count?</h2>
<p style= "color:#010f0f; font-family:Verdana,Arial">Reticulocytes are red blood cells that are still developing. They are also known as immature red blood cells.<br/> Reticulocytes are made in the bone marrow and sent into the bloodstream. About two days after they form, they develop into mature red blood cells.<br/> These red blood cells move oxygen from your lungs to every cell in your body.</p>
<p style= "color:#010f0f; font-family:Verdana,Arial">A reticulocyte count (retic count) measures the number of reticulocytes in the blood. If the count is too high or too low, it can mean a serious health problem, including anemia and disorders of the bone marrow, liver, and kidneys.</p>

<h3 style="color:#38bdc2; font-family:Verdana,Arial; font-size:15pt">II)What is it used for?</h3>
<p style= "color:#010f0f; font-family:Verdana,Arial">A reticulocyte count is most often used to:
    <ul>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Diagnose specific types of anemia. </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">See if treatment for anemia is working </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">See if bone marrow is producing the right amount of blood cells </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Check bone marrow function after chemotherapy or a bone marrow transplant</li>
    </ul></p>

<h4 style="color:#38bdc2; font-family:Verdana,Arial; font-size:15pt">III)What do the results mean?</h4>
<p style= "color:#010f0f; font-family:Verdana,Arial">If your results show a higher than normal amount of reticulocytes (reticulocytosis), it may mean:  
    <ul>
        <li style= "color:#010f0f; font-family:Verdana,Arial">You have hemolytic anemia, a type of anemia in which red blood cells are destroyed faster than the bone marrow can replace them. </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Your baby has hemolytic disease of the newborn, a condition that limits the ability of a baby's blood to carry oxygen to organs and tissues. </li>
    </ul></p>
<p style= "color:#010f0f; font-family:Verdana,Arial">If your results show a lower than normal amount of reticulocytes, it may mean:
    <ul>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Iron deficiency anemia, a type of anemia that happens when you don't have enough iron in your body </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Pernicious anemia, a type of anemia caused by not getting enough of certain B vitamins (B12 and folate) in your diet, or when your body can't absorb enough B vitamins. </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Aplastic anemia, a type of anemia that happens when the bone marrow isn't able to make enough blood cells.</li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Bone marrow failure, which may be caused by an infection or cancer.</li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Kidney disease</li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Cirrhosis, scarring of the liver</li>
    </ul></p>

<h1 style= "text-align:center;color:#052261;font-size:120%;"><u> LACTATE DEHYDROGENASE (LDH) TEST  </u></h2>
<h2 style= "color:#38bdc2; font-family:Verdana,Arial; font-size:15pt"> I)What is a lactate dehydrogenase (LDH) test?</h2>
<p style= "color:#010f0f; font-family:Verdana,Arial">This test measures the level of lactate dehydrogenase (LDH), also known as lactic acid dehydrogenase, in your blood or sometimes in other body fluids.<br/>  LDH is a type of protein, known as an enzyme. LDH plays an important role in making your body's energy.<br/> It is found in almost all the body's tissues, including those in the blood, heart, kidneys, brain, and lungs.</p>
<p style= "color:#010f0f; font-family:Verdana,Arial">When these tissues are damaged, they release LDH into the bloodstream or other body fluids. If your LDH blood or fluid levels are high, it may mean certain tissues in your body have been damaged by disease or injury</p>

<h3 style="color:#38bdc2; font-family:Verdana,Arial; font-size:15pt">II)What is it used for?</h3>
<p style= "color:#010f0f; font-family:Verdana,Arial">An LDH test is most often used to:
    <ul>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Find out if you have tissue damage </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Monitor disorders that cause tissue damage. These include anemia, liver disease, lung disease, and some types of infections. </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Monitor chemotherapy for certain types cancer. The test may show if treatment is working. </li>
    </ul></p>

<h4 style="color:#38bdc2; font-family:Verdana,Arial; font-size:15pt">III)What do the results mean?</h4>
<p style= "color:#010f0f; font-family:Verdana,Arial">Higher than normal LDH levels usually means you have some type of tissue damage or disease.<br/> Disorders that cause high LDH levels include:  
    <ul>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Anemia </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Kidney diseases </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Muscle injury </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Liver disease </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Myocardial infarction </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Pancreatitis </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Infections, including meningitis, encephalitis, and infectious mononucleosis (mono) </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Certain types of cancer, including lymphoma and leukemia. A higher than normal LDH level may also mean treatment for cancer is not working. </li>
    </ul></p>   
<p style= "color:#010f0f; font-family:Verdana,Arial">Although the test can show if you have tissue damage or disease, it does not show where the damage is located. If your results showed higher than normal LDH levels, your provider may need to order more tests to make a diagnosis. One of these tests may be an LDH isoenzyme test. An LDH isoenzyme test measures different forms of LDH. It can help your provider find out about the location, type, and severity of tissue damage<p/>

</body>

<h1 style= "text-align:center;color:#052261;font-size:120%;"><u> COOMBS TEST  </u></h2>
<h2 style= "color:#38bdc2; font-family:Verdana,Arial; font-size:15pt"> I)What is a Coombs test?</h2>
<p style= "color:#010f0f; font-family:Verdana,Arial">The Coombs test looks for antibodies that may stick to your red blood cells and cause red blood cells to die too early.</p>

<h3 style="color:#38bdc2; font-family:Verdana,Arial; font-size:15pt">II)Why the Test is Performed?</h3>
<p style= "color:#010f0f; font-family:Verdana,Arial">There are two types of the Coombs test: Direct and Indirect.
    <ul>
        <li style= "color:#010f0f; font-family:Verdana,Arial">The <strong>Direct Coombs test</strong> is used to detect antibodies that are stuck to the surface of red blood cells. Many diseases and drugs can cause this to happen. These antibodies sometimes destroy red blood cells and cause anemia. Your health care provider may recommend this test if you have signs or symptoms of anemia or jaundice (yellowing of the skin or eyes). </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">The <strong>indirect Coombs test</strong> looks for antibodies that are floating in the blood. These antibodies could act against certain red blood cells. This test is most often done to determine if you may have a reaction to a blood transfusion </li>
    </ul></p>

<h4 style="color:#38bdc2; font-family:Verdana,Arial; font-size:15pt">III)What do the results mean?</h4>
<p style= "color:#010f0f; font-family:Verdana,Arial"><strong>Normal Results:</strong><br/> A normal result is called a negative result. It means there was no clumping of cells and you have no antibodies to red blood cells.<br/><strong>Abnormal Results:</strong><br/>An abnormal (positive) direct Coombs test means you have antibodies that act against your red blood cells. This may be due to:  
    <ul>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Autoimmune hemolytic anemia </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Chronic lymphocytic leukemia or similar disorder </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Blood disease in newborns called erythroblastosis fetalis (also called hemolytic disease of the newborn) </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Infectious mononucleosis </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Mycoplasma infection </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Syphilis </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Systemic lupus erythematosus </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Transfusion reaction, such as one due to improperly matched units of blood </li>
    </ul></p>   
<p style= "color:#010f0f; font-family:Verdana,Arial">An abnormal (positive) indirect Coombs test means you have antibodies that will act against red blood cells that your body views as foreign. This may suggest:
    <ul>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Erythroblastosis fetalis </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Incompatible blood match (when used in blood banks) </li>
    </ul></p>
</body>
</html>

<h1 style= "text-align:center;color:#052261;font-size:120%;"><u> PERIPHERAL BLOOD SMEAR  </u></h2>
<p style= "color:#010f0f; font-family:Verdana,Arial"> A blood smear is a blood test that gives information about the number and shape of blood cells. It is often done as part of or along with a complete blood count (CBC)</p>    
  
<h2 style= "color:#38bdc2; font-family:Verdana,Arial; font-size:15pt"> I)How the Test is Performed?</h2>
<p style= "color:#010f0f; font-family:Verdana,Arial">A blood sample is needed.<br/>The blood sample is sent to a lab. There, the lab technician looks at it under a microscope. Or, the blood may be examined by an automated machine.<br/>The smear provides this information:
    <ul>
        <li style= "color:#010f0f; font-family:Verdana,Arial">The number,size(microcytic,normocytic or macrocytic) and kinds of red blood cells (differential, or percentage of each type of cell) </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">The number and kinds of abnormally shaped red blood cells</li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">A rough estimate of white blood cell and platelet counts.</li>
    </ul></p>

<h3 style="color:#38bdc2; font-family:Verdana,Arial; font-size:15pt">II)Why the Test is Performed?</h3>
<p style= "color:#010f0f; font-family:Verdana,Arial">This test may be done as part of a general health exam to help diagnose many illnesses. Or, your health care provider may recommend this test if you have signs of:
    <ul>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Any known or suspected blood disorder </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Cancer </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">Leukemia </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">A blood smear may also be done to monitor the side effects of chemotherapy or to help diagnose an infection, such as malaria </li>
    </ul></p>

        <li style= "color:#010f0f; font-family:Verdana,Arial">3+ means three quarters of cells are affected </li>
        <li style= "color:#010f0f; font-family:Verdana,Arial">4+ means all of the cells are affected </li>
    </ul></p>
