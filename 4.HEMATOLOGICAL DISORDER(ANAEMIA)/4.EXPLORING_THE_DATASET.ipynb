## <font color='#4287f5'>Feature_engineering...</font>

```python
# Lets define a function to convert Peripheral_Smear values in the integer datatype
def convert_Peripheral_Smear(x):
    if x == 'Microcytic':             
        return 7
    elif x == 'Normocytic':        
        return 8
    else:
        return 9                     
```

```python
df['Peripheral_Smear'] = df['Peripheral_Smear'].apply(lambda x: convert_Peripheral_Smear(x))
```

```python
df
```

```python
df['Coombs_test'].value_counts()
```

```python
# Lets define a function to convert Coombs_test values in the integer datatype
def convert_Coombs_test(x):
    if x == 'positive':              
        return 5
    elif x == 'negative':           
        return 6
    else:
        return x                     
```

```python
df['Coombs_test'] = df['Coombs_test'].apply(lambda x: convert_Coombs_test(x))
```

```python
df
```

```python
def convert_Other_blood_cell_count(x):
    if x == 'normal_count':              
        return 15
    else:
        return 14                     
```

```python
df['Other_blood_cell_count\(Pancytopenia +/-)'] = df['Other_blood_cell_count\(Pancytopenia +/-)'].apply(lambda x: convert_Other_blood_cell_count(x))
```

```python
df
```

```python
def convert_Hb_electrophoresis(x):
    if x == 'normal':              
        return 13
    elif x == 'HbA2,HbF +ve':
        return 12
    else:
        return x
```

```python
df['Hb_electrophoresis'] = df['Hb_electrophoresis'].apply(lambda x: convert_Hb_electrophoresis(x))
```
